```python
import matplotlib.pyplot as plt

# Total crimes per year
crimes_per_year = df.groupby('Financial Year')['Number of Offences'].sum().reset_index()
crimes_per_year.columns = ['Year', 'Total Offences']
print(crimes_per_year)

# Bar chart
plt.figure(figsize=(12,6))
plt.bar(crimes_per_year['Year'], crimes_per_year['Total Offences'], color='steelblue')
plt.title('Total Crimes Per Year (2012-2024)', fontsize=14)
plt.xlabel('Financial Year')
plt.ylabel('Number of Offences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 police forces
top_forces = df.groupby('Force Name')['Number of Offences'].sum()
top_forces = top_forces.sort_values(ascending=False).head(10)
print(top_forces)

# Horizontal bar chart
plt.figure(figsize=(12,6))
top_forces.sort_values().plot(kind='barh', color='coral')
plt.title('Top 10 Police Forces by Total Offences (2012-2024)', fontsize=14)
plt.xlabel('Total Number of Offences')
plt.tight_layout()
plt.show()
```

```python
# Most common crime types
crime_types = df.groupby('Offence Group')['Number of Offences'].sum()
crime_types = crime_types.sort_values(ascending=False)
print(crime_types)

# Pie chart
plt.figure(figsize=(12,8))
crime_types.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Crime Distribution by Offence Group (2012-2024)', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
```
